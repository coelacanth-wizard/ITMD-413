```python
import matplotlib.pyplot as plt

#open the file and read the lines into a list, but take out the title label lines first 
deptCount = open('employee_count_by_dept.txt', "r")
lines = deptCount.readlines()
lines.pop(0)

deptData = {} #make a dictionary to hold dept and count values
for line in lines: 
        # Split each line then store as dept:count pairs. Make sure the profit is being read as an int
        key, value = line.strip().split(', ') 
        deptData[key] = int(value) #don't need to strip number because they're in the hundreds

plt.pie(deptData.values(), labels = deptData.keys(), autopct='%1.1f%%')
plt.title("Employees per Department")
plt.show()

```

```python
import matplotlib.pyplot as plt

#open the file and read the lines into a list, but take out the title label lines first 
profit = open('last_ten_year_net_profit.txt', "r")
lines = profit.readlines()
label = lines[0].strip().split(';')
lines.pop(0)
profitData = {} #make a dictionary to hold year and profit values
for line in lines: 
        # Split each line then store as year:profit pairs. Make sure the profit is being read as an int
        key, value = line.strip().split(';') 
        profitData[key] = int(value.strip().replace(',', '')[1:])

plt.plot(profitData.keys(), profitData.values())

plt.xlabel(label[0])
plt.ylabel(label[1])

plt.title('Company profit in the past 10 Years')
plt.grid(True)
plt.show()

```
