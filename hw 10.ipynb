```python
# a) Using dataset workerstips.csv, create a scatterplot with total_bill as the x-axis and the tips as the y-axis.
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
print ("scatter plot...")

#only need the first and second values of each line 
workers = pd.read_csv('csv\workerstips.csv')

ax = sns.scatterplot(x='total_bill', y='tip', data = workers)
ax.set(xlabel ='Total Bill', ylabel ='Tips')

plt.title('Workers Tips')
```

```python
'''b) Using dataset workerstips.csv, create a scatter plot that will differentiate between smokers
    and non-smokers. Use a different color (hue) and style marker (style) to visually split the
    points of scatterplot. Because the default range sizes are limited, increase the range size.
    Try the tuple (10, 300) as the size range parameter.'''

ax = sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'smoker', style='smoker', palette='mako', size = 'size', sizes = (10,200), data = workers)
ax.set(xlabel = 'Total Bill', ylabel ='Tips')
plt.title('Workers Tips: Smokers and Non-Smokers')

```

```python
'''- c) A simple bar plot will work well to show average tips per day of the week. Find out
    which day has the highest average tip. In this case, you may want to show the average tip
    (y-axis) per day of the week (x-axis).'''
 
ax = sns.barplot(x = 'day', y = 'tip', order = ['Thur', 'Fri', 'Sat', 'Sun'], hue='day', palette='rocket', data = workers)
ax.set(xlabel = 'Day', ylabel = 'Average Tips')
plt.title('Average Tips per Day')
```

```python
'''- d) Repeat part (b) above but this time split the distribution by lunch and dinner. What
    information can you derive from the new visualization?'''

ax = sns.boxplot(x = 'day', y = 'total_bill', hue='time', palette='magma', order = ['Thur', 'Fri', 'Sat', 'Sun'], hue_order = ['Lunch', 'Dinner'], data = workers)
ax.set(xlabel = 'Day', ylabel ='Tips')
plt.title('Workers Tips: Lunch and Dinner')
```

```python
'''- e) Using dataset flightData.csv, create a lineplot to plot average number of passengers per
    year per month. Use the year on the x-axis and the passengers on the y-axis. Also, find
    out how to plot the total number of passengers per year.'''
flights = pd.read_csv('flightData.csv') #WHY DOESN'T RELATIVE PATH WORK. WHAT??? 
passPerYear = flights.groupby('year').sum()

# print(passPerYear)
ax = sns.lineplot(x = 'year', y = 'passengers', data = flights)
ax = sns.lineplot(x = 'year', y = 'passengers', data = passPerYear, dashes=True)

ax.set(xlabel = 'Year', ylabel = 'Passengers')
plt.title('Average Yearly and Total Monthly Passengers')
```
