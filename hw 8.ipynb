### A (PANDAS: SERIES) Perform the following tasks with pandas Series:
- a. Create a Series from the list [7, 11, 13, 17].
- b. Create a Series with five elements that are all 100.0.
- c. Create a Series with 20 elements that are all random numbers in the range 0 to 100. Use method describe to produce the Series’ basic descriptive statistics.
- d. Create a Series called temperatures of the floating-point values 98.6, 98.9, 100.2 and 97.9. Using the index keyword argument, specify the custom indices 'Julie', 'Charlie', 'Sam' and 'Andrea'.
- e. Form a dictionary from the names and values in Part (d), then use it to initialize a Series.

```python
import numpy as np
import pandas as pd
import random


ser = pd.Series([7, 11, 13, 17])
print("Series a.")
print(ser)

ser = pd.Series([100.0]*5)
print("\nSeries b.")
print(ser)

ser = pd.Series(np.random.randint(0, 100, size=20))
print("\nSeries c.")
print(ser.describe())

temperatures = pd.Series(data = [98.6, 98.9, 100.2, 97.9], index=['Julie', 'Charlie', 'Sam', 'Andrea'], name='Temperatures')
print("\nSeries d.")
print(temperatures)

tempDict = temperatures.to_dict()
print("\nSeries e.")
print(tempDict)
ser = pd.Series(data=tempDict)
print(ser)
```

### B (PANDAS: DATAFRAMES) Perform the following tasks with pandas DataFrames:
- f. Create a DataFrame named temperatures from a dictionary of three temperature
readings each for 'Maxine', 'James' and 'Amanda'.
- g. Recreate the DataFrame temperatures in Part (a) with custom indices using the index keyword argument and a list
containing 'Morning', 'Afternoon' and 'Evening'.
- h. Select from temperatures the column of temperature readings for 'Maxine'.
- i. Select from temperatures the row of 'Morning' temperature readings.
- j. Select from temperatures the rows for 'Morning' and 'Evening' temperature
readings.
- k. Select from temperatures the columns of temperature readings
for 'Amanda' and 'Maxine'.
- l. Select from temperatures the elements for 'Amanda' and 'Maxine' in
the 'Morning' and 'Afternoon'.
- m. Use the describe method to produce temperatures’ descriptive statistics.
- n. Transpose temperatures.
- o. Sort temperatures so that its column names are in alphabetical order.

```python
import numpy as np
import pandas as pd

d = {'Maxine': [53.4, 100.2, 59.6], 'James':[78.9, 98.9, 70.4], 'Amanda': [44.3, 98.6, 84.5]}

temperatures = pd.DataFrame(d)
print('DataFrame f.')
print(temperatures)

newIdx = ['Morning', 'Afternoon', 'Evening'] #this makes the random temps look sooo silly 
temperatures.index = newIdx
print('\nDataFrame g.')
print(temperatures)

print('\nDataFrame h.')
print(temperatures.get('Maxine'))

print('\nDataFrame i.')
print(temperatures[:1])

print('\nDataFrame j.')
print(temperatures.iloc[[0,-1]]) 

print('\nDataFrame k.')
print(temperatures.iloc[:,[0,-1]]) 

print('\nDataFrame l.')
print(temperatures.iloc[:2,[0,-1]]) 

print('\nDataFrame m.')
print(temperatures.describe())

print('\nDataFrame n.')
print(temperatures.transpose())

print('\nDataFrame o.')
print(temperatures.reindex(sorted(temperatures.columns), axis=1))
```

# Question 2
Using the titanic.csv dataset, write a program to explore and mine the following information:
- How many passengers were on the titanic?
- How many male and female passengers were in the titanic?
- What was the average age of all passengers?
- How many passengers were under 21 years of age?
- How many survived and how many did not? How many males and how many females?
- What was the youngest age that survived, and the oldest age? What were their names?
- Display the name of all passengers that survived

```python
import numpy as np
import pandas as pd

titanic = pd.read_csv('titanic.csv',)
# print(titanic)
print("\n\nThere were ",  len(titanic),"passengers on the Titanic.")
print("Number of Male Passengers:",  titanic['sex'].value_counts()['male'])
print("Number of Female Passengers:", titanic['sex'].value_counts()['female'] )

print("The average age of all the passengers was: ", titanic.loc[:, 'age'].mean())
print("There are ",  (titanic['age'] < 21).sum() ," passengers under 21 years old.")

print("Total Passengers Survived: ",  (titanic['survived'] == 'yes').sum()," Male: ", len(titanic.loc[(titanic['survived'] == 'yes') & (titanic['sex'] == 'male')]) ,", Female: ",  len(titanic.loc[(titanic['survived'] == 'yes') & (titanic['sex'] == 'female')]) )
print("Total Passengers Deceased: ",  (titanic['survived'] == 'no').sum()," Male: ", len(titanic.loc[(titanic['survived'] == 'no') & (titanic['sex'] == 'male')]) ,", Female: ",  len(titanic.loc[(titanic['survived'] == 'no') & (titanic['sex'] == 'female')]) )

young = titanic['age'].min()
youngeIdx = titanic.index.get_loc(titanic[titanic['age'] == young].index[0])

print("The youngest survivor was",  titanic['Name'][youngeIdx],". They were", young,"years old.")

oldest = titanic['age'].max()
oldestIdx = titanic.index.get_loc(titanic[titanic['age'] == oldest].index[0])
print("The oldest survivor was",  titanic['Name'][oldestIdx],"They were", oldest,"years old.")

print('\nList of passengers:\n',titanic['Name'])
```
