```python

from tkinter import *
from tkinter import messagebox
from tkinter import simpledialog
from tkinter.ttk import *

class Money():
    def __init__(self, Q, D, N, P, HDoll, Doll, Money = dict()): #short-handing this bc ball is life 
        self.__Quarter = Q
        self.__Dime = D
        self.__Nickel = N
        self.__Penny = P
        self.__HalfDollar = HDoll
        self.__Dollar = Doll
        # self.__Money = Money #TURNS OUT THIS WAS DUMB AND OVERCOMPLICATED. DISREGARD
    #     self.countMoney()
    
    # def countMoney(self):
    #     self.__Money.setdefault('Quarter', (self._Quarter, (self._Quarter * .25)))             
    #     self.__Money.setdefault('Dime', (self._Dime, (self._Dime * .1)))                       
    #     self.__Money.setdefault('Nickel', (self._Nickel, (self._Nickel * .05)))                
    #     self.__Money.setdefault('Penny', (self._Penny, (self._Penny * .01)))                   
    #     self.__Money.setdefault('Half-Dollar', (self._HalfDollar, (self._HalfDollar * .5)))        
    #     self.__Money.setdefault('Dollar', (self._Dollar, (self._Dollar * 1)))                  
        
    def totalMoney(self):
        return self._Quarter + self.__Dime + self.__Nickel + self._Penny + self._HalfDollar + self._Dollar
        
    @property
    def _Quarter(self):
        return self.__Quarter * .25

    @_Quarter.setter
    def _Quarter(self, value):
        self.__Quarter = value

    @property
    def _Dime(self):
        return self.__Dime * .1

    @_Dime.setter
    def _Dime(self, value):
        self.__Dime = value

    @property
    def _Nickel(self):
        return self.__Nickel * .05

    @_Nickel.setter
    def _Nickel(self, value):
        self.__Nickel = value

    @property
    def _Penny(self):
        return self.__Penny * .01

    @_Penny.setter
    def _Penny(self, value):
        self.__Penny = value

    @property
    def _HalfDollar(self):
        return self.__HalfDollar * .5

    @_HalfDollar.setter
    def _HalfDollar(self, value):
        self.__HalfDollar = value

    @property
    def _Dollar(self):
        return self.__Dollar

    @_Dollar.setter
    def _Dollar(self, value):
        self.__Dollar = value

win = Tk() #make the window
win.title('HW 12')
win.geometry('350x250')

welc = Label(win ,text = "Welcome to the program!").grid(row = 0,column = 0)

nm = Label(win, text = 'Quarters:').grid(column = 0, row = 1)
qrtr = Entry(win, width=15)
qrtr.grid(column = 1,row = 1)

nm = Label(win, text = 'Dimes:').grid(column = 0, row = 2)
dime = Entry(win, width=15)
dime.grid(column = 1,row = 2)

nm = Label(win, text = 'Nickels:').grid(column = 0, row = 3)
nick = Entry(win, width=15)
nick.grid(column = 1,row = 3)

nm = Label(win, text = 'Pennies:').grid(column = 0, row = 4)
penn = Entry(win, width=15)
penn.grid(column = 1,row = 4)

nm = Label(win, text = 'Half-Dollar:').grid(column = 0, row = 5)
half = Entry(win, width=15)
half.grid(column = 1,row = 5)

nm = Label(win, text = 'Dollar:').grid(column = 0, row = 6)
doll = Entry(win, width=15)
doll.grid(column = 1,row = 6)

messg = Label(win ,text = "Enter the amount of each \ncoin type then hit 'Compute':")
messg.grid(row = 7,column = 0, columnspan=3)

def compute():
    coins = Money(float(qrtr.get()), float(dime.get()), float(nick.get()), float(penn.get()), float(half.get()), float(doll.get()))
    qrtr.config(state='disabled')
    dime.config(state='disabled')
    nick.config(state='disabled')
    penn.config(state='disabled')
    half.config(state='disabled')
    doll.config(state='disabled')
    
    messg.config(text="Total")
    messg.grid(row = 7,column = 0, columnspan=1)
    
    nm = Label(win, text = '$' + str(coins._Quarter)).grid(column = 2, row = 1)
    nm = Label(win, text = '$' + str(coins._Dime)).grid(column = 2, row = 2)
    nm = Label(win, text = '$' + str(coins._Nickel)).grid(column = 2, row = 3)
    nm = Label(win, text = '$' + str(coins._Penny)).grid(column = 2, row = 4)
    nm = Label(win, text = '$' + str(coins._HalfDollar)).grid(column = 2, row = 5)
    nm = Label(win, text = '$' + str(coins._Dollar)).grid(column = 2, row = 6)
    
    tots = coins.totalMoney()
    total = Label(win, text='$'+ str(tots))
    total.grid(column = 2,row = 7)
    
    btn.destroy()
    
btn = Button(win, text = 'Compute', command=compute)
btn.grid(column=1,row=10)
win.mainloop()
```

Create a program called password_checker.py which is responsible for checking that the
password meets some pre-defined criteria. The following function is required to be implement
within this program:

• does_password_pass_check
> o Input parameter:
>> ▪ password -> string

> o Output parameter: 
>> ▪ Boolean value: True, if the password contains at least one digit (0 to 9)

>> ▪ Boolean value: False, if the password does not contain at least one digit (0 to 9) or an invalid value is entered as input
Note: This function must be a recursive function

• main
>o When this program is run the main function should be executed

>>o Call does_password_pass_check function with various valid and invalid cases/inputs and print the output

```python
def does_password_pass_check(pswd):
    if pswd[0].isdigit():
        return True
    elif len(pswd) <= 1 : 
        return False
    else: 
        return does_password_pass_check(pswd[1:])

def checkPass(pswd):
    if does_password_pass_check(pswd):
        print('This password DOES meet the necessary requirements!')
    else:
        print("This password DOES NOT meet the necessary requirements!")

def main():
    print('Welcome to the password checker!')
    pswd = input('Please enter a password: ')
    if does_password_pass_check(pswd):
        print('Your password DOES meet the necessary requirements!')
    else:
        print("This password DOES NOT meet the necessary requirements!")
        
    print('\n', '---- A series of tests will now run ---- ')
    print('\nPassword: lwksjad123')
    checkPass('lwksjad123')
    
    print('\nPassword: dlakwjljd')
    checkPass('dlakwjljd')
    
    print('\nPassword: qwlieu312ewjh')
    checkPass('qwlieu312ewjh')
    
    print('\nPassword: 2o3812uele')
    checkPass('2o3812uele')
    
    print('\nPassword: p92e0u')
    checkPass('p92e0u')

main()
```
