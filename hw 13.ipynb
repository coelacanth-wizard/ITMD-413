> b) There are three types of Machine Learning Algorithms; Supervised Learning,
Unsupervised Learning, and Reinforcement Learning? Describe each one of them and
provide some examples on each algorithm.

Supervised machine learning (ML) is when a machine is fed a series of labeled input/output data and uses these examples to predict outputs from new given inputs. This type of ML is used in a variety of situations including email filtering, price prediction, and voice recognition. 

Unlike supervised ML, Unsupervised ML is only given unlabeled input data, so the algorithm requires user input to piece together patterns and predict behavior. Common uses include grouping customer behavior patterns, analyzing weather patterns, and anomaly detection. 

Reinforcement ML, on the other hand, mimics the learning behaviors of humans by approaching data with a trial and error mindset. When a right decision is made, the machine is rewarded, and vice versa, learning from every result to make better predictions in the future.  Implementations include natural language processing, traffic control, and marketing.

```python
'''15.6
Go to NOAAa Climate at a Glance page and download the available time series 
data for the New York City average annual temperatures from 1895 through 
present. Reimplement the simple linear regression case study of 15.4 
using the average yearly temperature data. [...] 
How does the temperature trend compare to the average january high temperature?''' #It gets hotter

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

temp = pd.read_csv('ave_yearly_temp_nyc_1895-2017.csv')

x_train, x_test, y_train, y_test = train_test_split(temp.Date.values.reshape(-1, 1), temp.Value.values,)

regressor = LinearRegression()
regressor.fit(X=x_train, y=y_train)
pred = regressor.predict(x_test)

plt.scatter(x_train, y_train)
plt.plot(x_train, regressor.predict(x_train), color='red')
plt.ylim((45,65))

plt.title('Average Yearly NYC Temp.')
plt.xlabel('Year')
plt.ylabel('Temperature')
```
